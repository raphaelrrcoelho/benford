```python
df = pandas.read_csv("data.csv", sep=';', decimal=",")
df = df[df['VALOR']>=10]
df.shape
```

```python
batch_size = 1000
batch_list = []
data = df.copy()
while data.shape[0] > batch_size:
    rows = data.sample(batch_size, replace=False).index
    batch = data.loc[rows]
    batch_list.append(batch)
    data.drop(rows, inplace=True)
batch_list.append(data)
```

```python
benford_first = [log10(1 + 1./i) for i in range(1, 10)]
benford_second = []
for d in range(0,10):
    benford_second.append(sum([log10(1 + 1./(10*k+d)) for k in range(1, 10)]))
```

```python
for batch_number, batch in enumerate(batch_list):
    
    digit_count_first = Counter()
    digit_count_second = Counter()
    
    for value in  batch['VALOR'].values:
        digits = str(value).replace(".","")[0:2]
        first_digit = digits[0]
        if first_digit == '0':
            print(value)
        digit_count_first[first_digit] += 1
        
        try:
            second_digit = digits[1]
            digit_count_second[second_digit] += 1
        except:
            continue
        
    d1 = pandas.Series(list(digit_count_first.values()), index=list(digit_count_first.keys()))
    d1.sort_index(inplace=True)
    proportion1 = d1 / d1.sum()
    fig, ax = plt.subplots(figsize=(15,5))
    width = 0.3
    ax.bar(list(map(int,proportion1.index)), proportion1.values, width, color='r')
    ax.bar(list(map(lambda x: x+0.3, map(int,proportion1.index))), benford_first, width,  color='b')
    plt.title("First digit - batch {}".format(batch_number+1))
    plt.show()
    
    
    d2 = pandas.Series(list(digit_count_second.values()), index=list(digit_count_second.keys()))
    d2.sort_index(inplace=True)
    proportion2 = d2 / d2.sum()
    fig, ax = plt.subplots(figsize=(15,5))
    width = 0.3
    ax.bar(list(map(int,proportion2.index)), proportion2.values, width, color='r')
    ax.bar(list(map(lambda x: x+0.3, map(int,proportion2.index))), benford_second, width, color='b')
    plt.title("Second digit - batch {}".format(batch_number+1))
    plt.show()
    print("\n--------------------------------------------------------------------------------------------------\n")
```
